```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/content/unnati_phase1_data_revised_sup01.csv")
alert = data['Alert']
alert_unique=[]
```

```python
# No. of alerts
for i in alert:
  if (i not in alert_unique):
    alert_unique.append(i)
print(alert_unique)
```

```python
#No. of Vehicles
veh_no=[]
vehicle=data['Vehicle']
for j in vehicle:
  if (j not in veh_no):
    veh_no.append(j)
print(veh_no)
```

```python
# latitude
lat_c=0
lon_c=0
d1=1
d2=1
lat_unique=[]
lat_rep=[]
lon_unique=[]
lon_rep=[]
lat=data['Lat']
lon=data['Long']
D1={}
for i in lat:
  if i not in lat_unique:
    lat_unique.append(i)
  else:
    lat_rep.append(i)
    D1[i]=d1+1
print("Unique latitude:",len(lat_unique))
print("Repeated latitude:",len(lat_rep))
# print(D)
sorted_items = sorted(D1.items(), key=lambda x: x[1])
new_Dict = dict(sorted_items)
print(new_Dict)
```

```python
print("Range of Latitude:",min(lat),max(lon))
print("Range of Longitude:",min(lon),max(lon))
```

```python
D2={}
for j in lon:
  if j not in lon_unique:
    lon_unique.append(j)
  else:
    lon_rep.append(j)
    D2[j]=d2+1
print("Unique longitude:",len(lon_unique))
print("Repeated longitude:",len(lon_rep))
# print(D)
sorted_items = sorted(D2.items(), key=lambda x: x[1])
new_Dict2 = dict(sorted_items)
print(new_Dict2)
```
